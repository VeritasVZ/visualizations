```python
import altair as alt
import pandas as pd
import numpy as np



pd.set_option('display.max_columns',100)
pd.set_option('display.max_colwidth',-1)
```

```python
df = pd.read_csv('UCU_practice/hw1/GLB.Ts+dSST.csv', header = 1)
```

```python
df.head(3)
```

```python
import matplotlib.pyplot as plt

%matplotlib notebook
```

```python
fig, ax = plt.subplots(figsize = (12,5))


ax.scatter(df.Year, df['J-D'], color = '#FF6805', s=40, edgecolor='black', linewidth='0.5')
ax.set_xlim(1879,2019)
ax.set_facecolor('white')
ax.xaxis.set_ticks(np.arange(1880, 2020, 10))
ax.tick_params(axis='x', colors='grey')
ax.tick_params(axis='y', colors='grey')
ax.grid(color = 'grey', alpha = 0.3)
ax.plot(df.Year, [0]*len(df), color = 'black', lw = 0.5)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)

ax.text(1992, 0.05, 'HOTTER THAN THE\n1880-1899 AVERAGE', multialignment = 'right', fontname = 'Verdana')
ax.text(2008, -0.1, 'COLDER', multialignment = 'right', fontname = 'Verdana')

ax.text(1896, df['J-D'][df.Year==1904]-0.02, '1904', multialignment = 'right', fontname = 'Verdana')
ax.text(1945, df['J-D'][df.Year==1944]-0.02, '1944', multialignment = 'right', fontname = 'Verdana')
ax.text(1991, df['J-D'][df.Year==1998]-0.02, '1998', multialignment = 'right', fontname = 'Verdana')
ax.text(2015, df['J-D'][df.Year==2014]-0.02, '2014', multialignment = 'right', fontname = 'Verdana')
ax.text(2016, df['J-D'][df.Year==2015]-0.04, '2015', multialignment = 'right', fontname = 'Verdana')
ax.text(2017, df['J-D'][df.Year==2016]-0.02, '2016', multialignment = 'right', fontname = 'Verdana', weight = 'bold')
ax.text(2018, df['J-D'][df.Year==2017]-0.02, '2017', multialignment = 'right', fontname = 'Verdana')

ax.text(1880,0.9,'Annual Global Surface Temperature,\n Relative to Late 19th Century Average', multialignment = 'left',
        weight = 'bold')

# plt.show()
```
